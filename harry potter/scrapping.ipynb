```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd 
```

```python
url="https://www.hp-lexicon.org/characters"
r = requests.get(url)
htmlContent=r.content
```

```python
soup=BeautifulSoup(htmlContent,'html.parser')
#print(soup.prettify)
```

```python
# parent = soup.find("body").find("ul")
# print(parent)

# data1 = soup.find('ul')
# for li in data1.find_all("li"):
#     print(li.text, end=" ")
```

```python
ul = soup.select("ul.pagination")[0]
links = [a["href"] for a in ul.select("a[href]")]
text = [a.text for a in ul.select("a[title]")]
```

```python
link_list=[]
for i in range(len(links)):
    link_text="https://www.hp-lexicon.org"+links[i]
    link_list.append(link_text)
    #print(link_list[25])
    
link_list
```

```python
#links_str = ' '.join(map(str, links))
```

```python
data={
    'Letter':text,
    'Link':link_list
}
letter_df=pd.DataFrame(data)
```

```python
letter_df
```

looping through the letter links

```python


for i in letter_df.index:
    url = letter_df['Link'][i]
```

```python
character_list=[]

for i in range(len(link_list)):
        letter_url=link_list[i]
        r2 = requests.get(letter_url)
        htmlContent2=r2.content
        soup2=BeautifulSoup(htmlContent2,'html.parser')
        
        #print(letter_url)      
        
        character = soup2.find_all("span", {'itemprop': 'headline'})
        #print(character)
        
        
        letter_names=[]        
        for text in character:
                
                name=text.get_text()
                letter_names.append(name)
                
        character_list.append(letter_names)           
                

     
        
```

```python
character_list
```

```python
all_characters = []
for elem in character_list:
    for item in elem:
        all_characters.append(item)
```

```python
characters_df= pd.DataFrame(all_characters)
characters_df.columns=['character']
characters_df
```

```python
characters_df.to_csv('characters.csv',index=False)
```
